```python
import csv
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

Z1 = []
with open('fitness.txt', 'rt') as csvfile:
    fitness_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in fitness_reader:
        temp = []
        for el in row:
            temp.append(int(el))
        Z1.append(temp)
Z1 = np.array(Z1)
#print(Z)

Z2 = []
with open('fitness2.txt', 'rt') as csvfile:
    fitness_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in fitness_reader:
        temp = []
        for el in row:
            temp.append(int(el))
        Z2.append(temp)
Z2 = np.array(Z2)
#print(Z)

Z3 = []
with open('fitness3.txt', 'rt') as csvfile:
    fitness_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in fitness_reader:
        temp = []
        for el in row:
            temp.append(int(el))
        Z3.append(temp)
Z3 = np.array(Z3)
#print(Z)

        
```

```python
fig = plt.figure()
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(1, 3, 1, projection='3d')

X = np.arange(1, 11, 1)
Y = np.arange(1, 11, 1)
X, Y = np.meshgrid(X, Y)
surf = ax.plot_surface(X, Y, Z1, cmap=cm.gist_stern)
ax.view_init(elev=50., azim=120)

ax = fig.add_subplot(1, 3, 2, projection='3d')
X = np.arange(1, 11, 1)
Y = np.arange(1, 11, 1)
X, Y = np.meshgrid(X, Y)
surf = ax.plot_surface(X, Y, Z2, cmap=cm.gist_stern)
ax.view_init(elev=50., azim=120)

ax = fig.add_subplot(1, 3, 3, projection='3d')
X = np.arange(1, 11, 1)
Y = np.arange(1, 11, 1)
X, Y = np.meshgrid(X, Y)
surf = ax.plot_surface(X, Y, Z3, cmap=cm.gist_stern)
ax.view_init(elev=50., azim=120)


plt.show()
```
